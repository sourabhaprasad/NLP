Write a Python program for the following preprocessing of text in NLP:
● Tokenization
● Filtration
● Script Validation
● Stop Word Removal
● Stemming

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

nltk.download('punkt')
nltk.download('stopwords')

def preprocess_text(text):
    tokens = nltk.word_tokenize(text)
    tokens = [t for t in tokens if t.isalpha()]
    tokens = [t for t in tokens if re.fullmatch(r"[A-Za-z]+", t)]
    stop_words = set(stopwords.words('english'))
    tokens = [t for t in tokens if t.lower() not in stop_words]
    stemmer = PorterStemmer()
    return [stemmer.stem(t) for t in tokens]

text = "NLP preprocessing includes Tokenization, cleaning the text, removing stopwords, and more!"
processed = preprocess_text(text)
processed
```
