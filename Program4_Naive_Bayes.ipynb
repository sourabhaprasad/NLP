```python
from collections import Counter

docs = [
    ("fun couple love love".split(), "comedy"),
    ("fast furious shoot".split(), "action"),
    ("couple fly fast fun fun".split(), "comedy"),
    ("furious shoot shoot fun".split(), "action"),
    ("fly fast shoot love".split(), "action")
]

D = "fast couple shoot fly".split()

class_counts = Counter(cls for _, cls in docs)
total_docs = len(docs)
priors = {cls: class_counts[cls] / total_docs for cls in class_counts}

vocab = set()
word_counts = {"comedy": Counter(), "action": Counter()}
total_words = {"comedy": 0, "action": 0}

for words, cls in docs:
    vocab.update(words)
    word_counts[cls].update(words)
    total_words[cls] += len(words)

V = len(vocab)

def likelihood(w, cls):
    return (word_counts[cls][w] + 1) / (total_words[cls] + V)

def class_probability(cls):
    prob = priors[cls]
    for w in D:
        prob *= likelihood(w, cls)
    return prob

print("P(Comedy|D) =", class_probability("comedy"))
print("P(Action|D) =", class_probability("action"))
```
