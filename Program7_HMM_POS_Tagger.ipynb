```python
import numpy as np
import nltk
import ssl
from nltk.corpus import brown
from nltk.tag.hmm import HiddenMarkovModelTrainer

# Load tagged corpus
tagged_sents = brown.tagged_sents(tagset='universal')

# Train-test split (90% train, 10% test)
split = int(0.9 * len(tagged_sents))
train_sents = tagged_sents[:split]
test_sents = tagged_sents[split:]

print("Training sentences:", len(train_sents))
print("Testing sentences:", len(test_sents))

# Train HMM Tagger
trainer = HiddenMarkovModelTrainer()
hmm_tagger = trainer.train_supervised(train_sents)

# Evaluate accuracy
accuracy = hmm_tagger.accuracy(test_sents)
print("\nHMM Tagger Accuracy:", round(accuracy, 4))

# Demonstrate tagging
def tag_sentence(text):
    tokens = nltk.word_tokenize(text)
    return hmm_tagger.tag(tokens)

print("\nExample Tagged Sentence:")
print(tag_sentence("Machine learning models evolve rapidly."))
```
